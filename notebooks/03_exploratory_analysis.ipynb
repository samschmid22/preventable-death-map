# 03 Exploratory Analysis

## Objective

Explore relationships between 2015–2019 age-adjusted premature mortality and selected CDC PLACES 2020 county-level health predictors.

## Key Framing

This analysis examines county-level associations, not individual-level causation. Predictors are separated into behavioral/access indicators and disease-burden indicators because some variables, such as coronary heart disease and stroke prevalence, are closely related to mortality outcomes.

```python
import pandas as pd
from pathlib import Path

data_path = Path('../data/processed/mortality_places_merged_2015_2020.csv')
df = pd.read_csv(data_path, dtype={'county_fips': str})

df.shape
```

```python
target = 'premature_mortality_age_adjusted_rate_2015_2019'

behavioral_predictors = [
    'uninsured_rate',
    'smoking_rate',
    'obesity_rate',
    'physical_inactivity_rate',
    'poor_physical_health_rate',
    'poor_mental_health_rate',
    'insufficient_sleep_rate'
]

disease_burden_indicators = [
    'diabetes_rate',
    'high_blood_pressure_rate',
    'copd_rate',
    'coronary_heart_disease_rate',
    'stroke_rate'
]

all_predictors = behavioral_predictors + disease_burden_indicators

analysis_df = df[[target] + all_predictors].dropna()
analysis_df.shape
```

```python
correlations = (
    analysis_df
    .corr(numeric_only=True)[target]
    .drop(target)
    .sort_values(ascending=False)
)

correlations
```

```python
behavioral_correlations = correlations.loc[behavioral_predictors].sort_values(ascending=False)
disease_burden_correlations = correlations.loc[disease_burden_indicators].sort_values(ascending=False)

behavioral_correlations, disease_burden_correlations
```

```python
top_10_mortality = (
    df[[
        'county_fips',
        'county',
        target,
        'smoking_rate',
        'obesity_rate',
        'physical_inactivity_rate',
        'diabetes_rate'
    ]]
    .dropna(subset=[target])
    .sort_values(target, ascending=False)
    .head(10)
)

top_10_mortality
```

## Initial Takeaways

- Premature mortality shows strong positive county-level correlations with multiple health indicators.
- Disease-burden indicators such as coronary heart disease, stroke, COPD, and high blood pressure have especially strong associations with premature mortality.
- Behavioral and health-status indicators such as smoking, physical inactivity, poor physical health, and poor mental health also show strong associations.
- Future modeling should separate upstream predictors from disease-burden indicators to avoid making the analysis feel circular.
